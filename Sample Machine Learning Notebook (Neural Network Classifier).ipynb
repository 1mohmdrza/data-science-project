# Step1: Define X & Y

# Step3: Initiallize Neural Network

```python
from sklearn.neural_network import MLPClassifier
```

```python
MLP = MLPClassifier()
```

# Step4: Fit & Measure

```python
from sklearn.model_selection import cross_val_predict , cross_val_score
```

```python
scores = cross_val_score(MLP, X , Y , cv=10 , scoring='accuracy')
scores.mean()
```

```python
MLP = MLPClassifier(activation='relu')
scores = cross_val_score(MLP, X , Y , cv=10 , scoring='accuracy')
scores.mean()
```

```python
MLP = MLPClassifier(activation='relu' , hidden_layer_sizes =(100,120),
                    learning_rate_init=0.05,
                    random_state=8)
scores = cross_val_score(MLP, X , Y , cv=10 , scoring='accuracy')
scores.mean()
```

```python
MLP = MLPClassifier(activation='relu' , hidden_layer_sizes =(100,120),
                    learning_rate_init=0.05, learning_rate='invscaling',
                    random_state=8)
scores = cross_val_score(MLP, X , Y , cv=10 , scoring='accuracy')
scores.mean()
```

```python
MLP = MLPClassifier(activation='relu' , hidden_layer_sizes =(100,120),
                    learning_rate_init=0.05, learning_rate='invscaling',
                    momentum=0.9 , random_state=8)
scores = cross_val_score(MLP, X , Y , cv=10 , scoring='accuracy')
scores.mean()
```

```python
MLP = MLPClassifier(activation='relu' , hidden_layer_sizes =(100,120),
                    learning_rate_init=0.05, learning_rate='invscaling',
                    momentum=0.9 , max_iter=50 , random_state=8)
scores = cross_val_score(MLP, X , Y , cv=10 , scoring='accuracy')
scores.mean()
```

```python
MLP = MLPClassifier(activation='relu' , hidden_layer_sizes =(100,120),
                    learning_rate_init=0.05, learning_rate='invscaling',
                    momentum=0.9 , max_iter=50 , verbose=1 , random_state=8)
scores = cross_val_score(MLP, X , Y , cv=10 , scoring='accuracy')
scores.mean()
```

# Grid Search

```python
from sklearn.model_selection import GridSearchCV
MLP = MLPClassifier()
```

```python
params = {'activation':['relu' , 'logestic' , 'tanh'],
         'hidden_layer_sizes':[(10), (20) , (20,30)],
         'max_iter':[10 , 50 , 100 , 200],
         'solver':['sgd' , 'adam'],
         'learning_rate_init':[0.01 , 0.001 , 0.0001 , 0.025]}

GS = GridSearchCV(MLP , params , cv=10 , scoring='accuracy')
GS.fit(X,Y)
```

# Validation 

```python
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score , confusion_matrix
from sklearn.neural_network import MLPClassifier

MLP = MLPClassifier(activation= 'tanh', hidden_layer_sizes= 20,
                    learning_rate_init= 0.01, max_iter= 200, solver= 'adam')
MLP.fit(X , Y)
Ypred = MLP.predict(X)
```

```python
acc = accuracy_score(Y , Ypred)
rec = recall_score(Y , Ypred)
pre = precision_score(Y , Ypred)
fm = f1_score(Y , Ypred)
cm = confusion_matrix(Y , Ypred)

print(acc, rec , pre , fm)
print(cm)
```
